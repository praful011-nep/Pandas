**Data Frames Applying Functions**

*.apply(function_name)*  
To use a user-defined function in the dataframe

```python
df
```

```python
def square(x):
    return x**2
```

```python
df['B'].apply(square)
```

The changes can also be updated in the main dataframe.  
Similarly a new column can be also created.

```python
df['D'] = df['B'].apply(square)
```

```python
df
```

lamda functions can also be used

```python
df['C'].apply(lambda x: x + 100)
```
