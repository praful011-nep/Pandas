**Data Frame**

```python
import pandas as pd
import numpy as np
```

*Creating a Dataframe*

Dictionary  
Creating dataframe from a dictionary, key becomes the header

```python
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [28, 34, 29, 42],
    'City': ['New York', 'Paris', 'Berlin', 'London'],
    'Salary': [65000, 70000, 62000, 85000]
}
df=pd.DataFrame(data)
df
```

List  
Creating dataframe from a list, header are default set as index  
For header, columns=*list having column title* is added in the function

```python
data_list = [
    ['John', 28, 'New York', 65000],
    ['Anna', 34, 'Paris', 70000],
    ['Peter', 29, 'Berlin', 62000],
    ['Linda', 42, 'London', 85000]
]
df2=pd.DataFrame(data_list)
columns=["Name","Age","City","Salary"]
df2=pd.DataFrame(data_list,columns=columns)
df2
```

*Selecting and Indexing of columns*

For one column a single ["*Column header*"] is required but for more list of list is required

```python
# df2["Name"]
df2[["Name","Age"]]
```

*Creating a new Columns*

*Dataframe*["*Column_header* "] = *list containing values*

```python
df2["Designation"]=["Doctor","Eng","Doctor","Eng"]
df2
```

*Removing Columns or Rows*

*Dataframe*.drop("*Column_name or Row_label*",axis=....)

axis is 0 in default so so axis should be given 1 to specify find the header of columns

```python
df2.drop("Designation", axis=1)
```

drop() just makes a copy of the dataframe rather than doing operations on the actual dataframe  
drop("*Column_name*",axis=1,inplace = True) is used for removing columns in the main data

```python
df2
```

```python
df2.drop("Designation", axis=1, inplace= True)
```

```python
df2.drop(0)
```
