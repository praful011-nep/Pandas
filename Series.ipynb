**Pandas Series**  
A series is a  one-dimensional labeled array capable of holding any datatype. The axis label are collectively called index.

```python
import numpy as np
import pandas as pd
```

**Creating Series**

```python
labels=['a','b','c']
my_list=[10,20,30]
arr=np.array([10,20,30])
d={1:10,2:20,3:30}
```

```python
pd.Series(my_list)
```

```python
#For custom labels
pd.Series(my_list,index=labels)
```

```python
#In dictionary, Keys becomes labels and Values become elements
pd.Series(d)
```
