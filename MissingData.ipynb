```python
import pandas as pd
import numpy as np
```

**Finding Missing Data**

```python
#np.nan means null value (Not a Number)
data = {
    'A': [1, 2, np.nan, 4, 5],        
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df=pd.DataFrame(data)
```

```python
df
```

*.isna()* : gives True if NaN  
.sum() gives the total number  
.any() gives is there a null value in whole column

```python
df.isna()
```

```python
df.isna().sum()
```

```python
df.isna().any() 
```

**Removing Missing Data**

```python
df
```

*.dropna()* removes all the rows with null value   
thresh means minimum number of null values to be dropped

```python
df.dropna()
```

```python
df
```

```python
df.dropna(thresh=3); 
```

**Filling the Missing Data**

Finding the missing data and filling data.

*fillna(data)* : fills *data* in place of null value  
*fillna(value= )*: a dictionary can be passed with column_header as key and replacing data as value

```python
df
```

```python
df.fillna(0)
```

```python
values={
    "A":0,
    "B": 100,
    "C":150,
    "D":120
}
df.fillna(value=values)
```

```python
df
```

```python
df.fillna(df.mean()) # fill mean in place of null data
```
